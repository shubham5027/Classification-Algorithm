# Softmx Regression or multinomial regression for predicting multiple classified labels

```python
# softmx regression or mulnomial regression for predicting multiple classified labels
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.linear_model import LogisticRegression
```

```python
df=sns.load_dataset('iris')
```

```python
df
```

```python
df.info()
```

```python
lb=LabelEncoder()
```

```python
df['species']=lb.fit_transform(df['species'])
```

```python
df.head()
```

```python
X=df.drop(['species'],axis=1)
y=df['species']
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=0,test_size=0.3)
```

SOFTMAX Regression also called Multinomial classification

```python
lr=LogisticRegression(multi_class='multinomial',solver='sag')
```

```python
lr.fit(X_train,y_train)
y_pred=lr.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test,y_pred))
print(accuracy_score(y_test,y_pred))
```
